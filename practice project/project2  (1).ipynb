```python
# from the data I understood that its a binary possibility problem of 0 & 1
# also missing values are there hence we will replace them with the median of Age and I feel cabin columns doesnt have any direct relation hence it can be dropped.
```

```python
df['Age'].fillna(df['Age'].median(),inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.drop(columns=['Cabin'], inplace=True)
```

```python
df.drop(columns=['Name'], inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(10, 8))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
sns.pairplot(df, hue='Survived', diag_kind='kde')
plt.show()
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)

plt.title('Survival Count by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
sns.barplot(x='Sex', y='Survived', data=df, ci=None)

# Set plot title and labels
plt.title('Survival Rate by Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')

# Show the plot
plt.show()
```

```python
df_ohe = pd.get_dummies(df, columns=['Sex', 'Embarked'], drop_first=True)
```

```python
df_ohe.drop(columns=['Ticket'], inplace=True)
```

```python
df_ohe.isnull().sum()
```

```python
X = df_ohe.drop(columns=['Survived'])
y = df_ohe['Survived'] 
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(X_train, y_train)
```

```python
y_pred = dtc.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
confusion = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)
```

```python
print(f"Accuracy: {accuracy}")
print("Confusion Matrix:\n", confusion)
print("Classification Report:\n", classification_rep)
```
