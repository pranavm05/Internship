Calculating mode and average for each source so as to fill the missing values in the year 2014,2015,2016,2017,2018

```python
mode_generation = df.groupby('primary_fuel')[['generation_gwh_2015', 'generation_gwh_2016', 'generation_gwh_2017', 'generation_gwh_2018']].agg(lambda x: x.mode().values[0] if not x.mode().empty else np.nan)
```

```python
mode_generation
```

```python
for year in ['2015', '2016', '2017', '2018']:
    column_name = 'generation_gwh_' + year
    is_missing = df[column_name].isnull()
    df.loc[is_missing, column_name] = df.loc[is_missing, 'primary_fuel'].map(mode_generation[column_name])
```

```python
mode_generation
```

```python
fuel_average = df.groupby('primary_fuel')['generation_gwh_2015', 'generation_gwh_2016', 'generation_gwh_2017', 'generation_gwh_2018'].transform('mean')
```

```python
fuel_average
```

```python
average_generation_2015 = df['generation_gwh_2015'].mean()
average_generation_2016 = df['generation_gwh_2016'].mean()
average_generation_2017 = df['generation_gwh_2017'].mean()
average_generation_2018 = df['generation_gwh_2018'].mean()

# Print the results
print("Average Generation (2015):", average_generation_2015)
print("Average Generation (2016):", average_generation_2016)
print("Average Generation (2017):", average_generation_2017)
print("Average Generation (2018):", average_generation_2018)
```

```python
# Fill null values in the specified columns with the calculated average values
df['generation_gwh_2015'].fillna(average_generation_2015, inplace=True)
df['generation_gwh_2016'].fillna(average_generation_2016, inplace=True)
df['generation_gwh_2017'].fillna(average_generation_2017, inplace=True)
df['generation_gwh_2018'].fillna(average_generation_2018, inplace=True)
```

```python
df
```

```python
average_generation_2014 = df['generation_gwh_2014'].mean()
print("Average Generation (2014):", average_generation_2014)
```

```python
df['generation_gwh_2014'].fillna(average_generation_2014, inplace=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
tree_classifier = DecisionTreeClassifier(random_state=42)
```

```python
tree_classifier.fit(X_train, y_train)
```

```python
y_pred = tree_classifier.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
confusion = confusion_matrix(y_test, y_pred)
report = classification_report(y_test, y_pred)

print("Accuracy:", accuracy)
print("Confusion Matrix:\n", confusion)
print("Classification Report:\n", report)





```

```python
from sklearn.model_selection import GridSearchCV
```

```python
tree_classifier = DecisionTreeClassifier(random_state=42)
```

```python
param_grid = {
    'max_depth': [None, 5, 10, 15],  # Maximum depth of the tree
    'min_samples_split': [2, 5, 10],  # Minimum samples required to split an internal node
    'min_samples_leaf': [1, 2, 4]  # Minimum samples required for a leaf node
}
```

```python
grid_search = GridSearchCV(estimator=tree_classifier, param_grid=param_grid, cv=5, scoring='accuracy')
```

```python
grid_search.fit(X_train, y_train)
```

```python
best_params = grid_search.best_params_
```

```python
tuned_tree_classifier = DecisionTreeClassifier(**best_params)
```

```python
tuned_tree_classifier.fit(X_train, y_train)
```

```python
y_pred = tuned_tree_classifier.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
confusion = confusion_matrix(y_test, y_pred)
report = classification_report(y_test, y_pred)

print("Best Hyperparameters:", best_params)
print("Accuracy:", accuracy)
print("Confusion Matrix:\n", confusion)
print("Classification Report:\n", report)
```
