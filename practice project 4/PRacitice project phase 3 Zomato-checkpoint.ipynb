from the unique values of Column average cost for two we can find that there are minimu values and then there are exorbitantly high values 150000, 70000 this values may cause bias or too much variance.

```python
Q1 = df['Average Cost for two'].quantile(0.25)
Q3 = df['Average Cost for two'].quantile(0.75)
```

```python
Q1
```

```python
Q3
```

```python
IQR = Q3 - Q1
```

```python
IQR
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
lower_bound
```

```python
upper_bound
```

```python
outliers = df[(df['Average Cost for two'] < lower_bound) | (df['Average Cost for two'] > upper_bound)]
```

```python
print("Identified outliers:")
print(outliers)
```

```python
cleaned_df = df[(df['Average Cost for two'] >= lower_bound) & (df['Average Cost for two'] <= upper_bound)]
```

```python
print(cleaned_df)
```

```python
df = cleaned_df
```
